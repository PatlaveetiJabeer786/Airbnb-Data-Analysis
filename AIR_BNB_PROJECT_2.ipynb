# 5️⃣ How has the number of reviews changed over time?

```python
df.head(1)
```

```python
print(df["last review"].isna().sum())
print(df["last review"].dropna().shape)
```

```python
df["last review"] = pd.to_datetime(df["last review"], errors='coerce')
```

```python
df_clean = df.dropna(subset=["last review"])
reviews_over_time = df_clean.groupby(df_clean["last review"].dt.to_period("M")).size()
```

```python
if not reviews_over_time.empty:
    plt.figure(figsize=(12,6))
    reviews_over_time.plot(kind="line", color="red")
    plt.title("Reviews Over Time")
    plt.ylabel("Number of Reviews")
    plt.xlabel("Month")
    plt.show()
else:
    print("No review data available to plot.")
```
